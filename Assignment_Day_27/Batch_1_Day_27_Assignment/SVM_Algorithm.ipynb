```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
df=pd.read_csv("G:/study_vidya/Assignments_LetsUpgrade/Batch_1_Day_27_Assignment/titanic_train_dataset.csv")
df.head()
```

```python
df = df.drop(['Name','PassengerId','Ticket','Cabin'],axis=1)
```

```python
le = LabelEncoder()
df['Sex'] = le.fit_transform(df['Sex'])
df['Embarked'] = le.fit_transform(df['Embarked'])
```

```python
df.head()
```

```python
def Model(col):
    x = df.drop([col],axis=1)
    y = df[col]
    train_x, test_x, train_y, test_y = train_test_split(x, y, test_size=0.25, random_state=42)
    SVM = SVC(gamma=0.01, C=100)
    SVM.fit(train_x, train_y)
    print(f"Accuracy score is {accuracy_score(SVM.predict(test_x),test_y, normalize=True)}")
    print("------------------------------------")
    print(f"Confusion matrix: \n{confusion_matrix(SVM.predict(test_x), test_y)}")
```

```python
Model('Pclass')
```
