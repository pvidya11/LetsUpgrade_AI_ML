```python
#COrrleation between Attrition and Age
from scipy.stats import pearsonr
stats,p=pearsonr(df_numerical.Attrition_Yes,dataset1.Age)
print(stats,p)
#H0:There is no significance difference between attrition with Age.
if p<0.05:
    print("Reject null hypothesis")
else:    
    print("Accept null hypothesis")
```
